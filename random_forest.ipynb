# Random Forest

Para comparar com o XGBoost, criamos um modelo de Random Forest, entretato o modelo apresentou um resultado pior.

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import  GridSearchCV
from sklearn.ensemble import RandomForestRegressor

%matplotlib inline
%store -r dataset_modelo
%store -r X_train
%store -r X_val
%store -r Y_train
%store -r Y_val
```

```python
params = {
    'n_estimators' : [100, 200],
    'max_depth' : [2,3, 4, 6]
}
rfreg = GridSearchCV(RandomForestRegressor(random_state=0), params)
rfreg.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error

Y_pred = rfreg.predict(X_val)
rmae = mean_absolute_error(y_pred=Y_pred, y_true=Y_val)

rmae
```
