# Random Forest

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import  GridSearchCV
from sklearn.ensemble import RandomForestRegressor

%matplotlib inline
%store -r dataset_modelo
%store -r X_train
%store -r X_val
%store -r Y_train
%store -r Y_val
```

```python
params = {
    'n_estimators' : [100, 200, 300, 400, 500, 800, 1000],
    'max_depth' : [2,3, 4, 6, 8, 10]  
}
rfreg = GridSearchCV(RandomForestRegressor(random_state=0), params)
rfreg.fit(X_train, Y_train)
```

```python
Y_pred = rfreg.predict(X_val)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

rmse = mean_squared_error(y_pred=Y_pred, y_true=Y_val)
rmae = mean_absolute_error(y_pred=Y_pred, y_true=Y_val)

print(rmae)
```

```python
print("RMSE : {:.2f}".format(rmse))
# %%
from sklearn.metrics import r2_score
r2_score(Y_pred, Y_val)
```
